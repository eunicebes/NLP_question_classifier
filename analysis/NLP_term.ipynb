```python
import json

class Members(object):
    """docstring for Members"""
    def __init__(self, datafile):
        super(Members, self).__init__()
        self.datafile = datafile

        # load member data
        with open(self.datafile) as jsonfile:
            self.data = json.load(jsonfile)

    def get_member_info_by_id(self, memberid):
        return self.data[memberid]

    def get_all_questions(self):
        questions = []
        for key, vals in self.data.items():
            for q in vals['question']:
                questions.append(q['message'])
        return questions
    
    # get the question description
    def get_questions(self):
        for key, vals in self.data.items():
            for q in vals['question']:
                yield q['message']
                
    def get_question_info(self):
        for key, vals in self.data.items():
            for q in vals['question']:
                yield q
    
    # get the description of question and corresponding replies
    def get_question_reply(self):
        for key, vals in self.data.items():
            for q in vals['question']:
                yield (q['message'], [rel['message'] for rel in q['reply'] ])

    def get_questions_count(self):
        count = 0
        for key, vals in self.data.items():
            for q in vals['question']:
                count += 1
        return count
```

## Read questions and replies into file

```python
# PT： pronounciation and translation
Grammar = {'完成式': 1, '進行式': 2, '過去式': 3, '未來式': 4, '關係代名詞': 5, '不定詞': 6, '名詞子句': 7, 
           '被動': 8, '介係詞': 9, '連接詞': 10, '假設語氣': 11, '現在分詞': 12, '過去分詞': 13, 'PT': 14, '其它': 0}
Grammar_cat = ['完成式', '進行式', '過去式', '未來式', '關係代名詞', '不定詞', '名詞子句', '被動', '介係詞', '連接詞', '假設語氣',\
              '現在分詞', '過去分詞']
```

```python
import csv
from collections import defaultdict, Counter
from progressbar import AnimatedMarker, Bar, BouncingBar, ETA, FileTransferSpeed, FormatLabel, Percentage, \
    ProgressBar, ReverseBar, RotatingMarker, \
    SimpleProgress, Timer

outfile = 'questions.csv'
question_count = member.get_questions_count()
question_dict = Counter()
with open(outfile, 'w') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter = ',', quotechar = '"')
    spamwriter.writerow(['question_id', 'member_id', 'question', 'type'])
    pbar  = ProgressBar(widgets=[Percentage(), Bar()], maxval = question_count).start()
    index = 0
    for question in member.get_question_info():
        q = question['message']
        replies = question['reply']
        
        # pronouciation and translation
        if '發音' in q or '聽' in q or '翻' in q:
            spamwriter.writerow([question['id'], question['memberid'], q, Grammar['PT']])
            question_dict['PT'] += 1
            continue
        
        for query in Grammar_cat:
            # if keyword is in question, this question is related to this grammar
            if query in q:
                spamwriter.writerow([question['id'], question['memberid'], q, Grammar[query]])
                question_dict[query] += 1
            else:
                for r in replies:
                    if query in r['message']:
                        spamwriter.writerow([question['id'], question['memberid'], q, Grammar[query]])
                        question_dict[query] += 1
                        break
        pbar.update(index+1)
        index += 1
    pbar.finish()
print('{} questions are processed.'.format(index-1))
pprint.pprint(question_dict)
```

## Define question class

```python
from collections import defaultdict, Counter
class Questions(Members):
    """docstring for Questions"""
    def __init__(self, class_A, datafile):
        if class_A.data != None:
            self.data = class_A.data
        else:
            # load member data
            with open(self.datafile) as jsonfile:
                self.data = json.load(jsonfile)
    
    def get_question_dict(self):
        questions = defaultdict(defaultdict)
        for q in self.get_question_info():
            questions[q['id']]['memberid'] = q['memberid']
            questions[q['id']]['message'] = q['message']
            questions[q['id']]['reply'] = [reply['message'] for reply in q['reply']]
        return questions
```

## Merge question and corresponding replies

```python
import csv
infile = 'questions_nondup_dup.csv'
outfile ='questions_nondup_dup2.csv'
with open(outfile, 'w') as outfile:
    spamwriter = csv.writer(outfile, delimiter = ',', quotechar = '"')
    spamwriter.writerow(['question_id', 'member_id', 'question', 'type', 'ambiguous'])
    
    with open(infile, 'r') as csvfile:
        for row in csv.DictReader(csvfile):
            question_id = row['question_id']
            question = row['question']
            replies = question_dict[question_id]['reply']
            reply = ""
            for rep in replies:
                reply += rep
            question += reply # merge
            writestring = [question_id, row['member_id'], question, row['type'], row['ambiguous']]
            spamwriter.writerow(writestring)
```

```python
import json
import csv
def conver2json(infile_unamb, infile_amb, outfile):
    predict_dict = defaultdict(dict)

    predict_dict = defaultdict(dict)
    with open(infile_unamb, 'r') as csvfile:
        for row in csv.DictReader(csvfile):
            q_id = row['question_id']
            replies = question_dict[q_id]['reply']
            reply = []
            reply_str = ''
            for rep in replies:
                reply.append(rep)
            reply_str = ' '.join(reply)
            
            predict_dict[q_id]['member_id'] = row['member_id']
            predict_dict[q_id]['question'] = row['question']
            predict_dict[q_id]['reply'] = reply_str
            predict_dict[q_id]['其它'] = row['其它']
            predict_dict[q_id]['完成式'] = row['完成式']
            predict_dict[q_id]['連接詞'] = row['連接詞']
            predict_dict[q_id]['假設語氣'] = row['假設語氣']
            predict_dict[q_id]['分詞'] = row['分詞']
            predict_dict[q_id]['進行式'] = row['進行式']
            predict_dict[q_id]['過去式'] = row['過去式']
            predict_dict[q_id]['未來式'] = row['未來式']
            predict_dict[q_id]['關係代名詞'] = row['關係代名詞']
            predict_dict[q_id]['不定詞'] = row['不定詞']
            predict_dict[q_id]['名詞子句'] = row['名詞子句']
            predict_dict[q_id]['被動'] = row['被動']
            predict_dict[q_id]['介係詞'] = row['介係詞']
            predict_dict[q_id]['ambiguous'] = 0
    
    with open(infile_amb, 'r') as csvfile:
        for row in csv.DictReader(csvfile):
            q_id = row['question_id']
            replies = question_dict[q_id]['reply']
            reply = []
            for rep in replies:
                reply.append(rep)
            reply_str = ' '.join(reply)
            
            predict_dict[q_id]['member_id'] = row['member_id']
            predict_dict[q_id]['question'] = row['question']
            predict_dict[q_id]['reply'] = reply_str
            predict_dict[q_id]['其它'] = row['其它']
            predict_dict[q_id]['完成式'] = row['完成式']
            predict_dict[q_id]['連接詞'] = row['連接詞']
            predict_dict[q_id]['假設語氣'] = row['假設語氣']
            predict_dict[q_id]['分詞'] = row['分詞']
            predict_dict[q_id]['進行式'] = row['進行式']
            predict_dict[q_id]['過去式'] = row['過去式']
            predict_dict[q_id]['未來式'] = row['未來式']
            predict_dict[q_id]['關係代名詞'] = row['關係代名詞']
            predict_dict[q_id]['不定詞'] = row['不定詞']
            predict_dict[q_id]['名詞子句'] = row['名詞子句']
            predict_dict[q_id]['被動'] = row['被動']
            predict_dict[q_id]['介係詞'] = row['介係詞']
            predict_dict[q_id]['ambiguous'] = 1
            
    with open(outfile, 'w') as jsonfile:
        json.dump(predict_dict, jsonfile)
```

## Generate student dict

```python
def covert2student(infile_unamb, infile_amb, outfile):
    student_dict = defaultdict(Counter)
    with open(infile_unamb, 'r') as csvfile:
        spamreader = csv.reader(csvfile, delimiter = ',', quotechar = '"')
        header = next(spamreader)
        classes = header[3:]
        for row in spamreader:
            prob = [float(p) for p in row[3:]]
            member_id = row[1]
            question_id = row[0]
            cls_name = classes[prob.index(max(prob))]
            student_dict[member_id][cls_name] += 1
    
    with open(infile_amb, 'r') as csvfile:
        spamreader = csv.reader(csvfile, delimiter = ',', quotechar = '"')
        header = next(spamreader)
        classes = header[3:]
        for row in spamreader:
            prob = [float(p) for p in row[3:]]
            member_id = row[1]
            question_id = row[0]
            cls_name = classes[prob.index(max(prob))]
            student_dict[member_id][cls_name] += 1
            
    with open(outfile, 'w') as jsonfile:
        json.dump(student_dict, jsonfile)
    
    return student_dict
```

```python
infile_unamb = 'predicted/LR_unamb_predict.csv'
infile_amb = 'predicted/LR_amb_predict.csv'
outfile = 'predicted/LR_student.json'
student_dict = covert2student(infile_unamb, infile_amb, outfile)
```
